# Data cleaning

1. normalized-losses  ->replace by mean

2. bore               ->Replace by mean

3. stroke             ->Reaplce by mean 

4. horsepower         ->Replace by mean

5. peak-rpm           ->Reaplce by mean 

4. num-of-doors       ->replace them with four Reason: 84% sedans is four doors. Since four doors is most frequent,                             it is most likely to occur.


```python
avg_norm_loss=df['normalized-losses'].astype('float').mean(axis=0)
print('Average of normalized-losses',avg_norm_loss)
```

```python
df['normalized-losses'].replace(np.nan,avg_norm_loss,inplace=True)
```

```python
avg_broe=df['bore'].astype('float').mean(axis=0)
print("Avrage Broke ",avg_broe)
```

```python
df['bore'].replace(np.nan,avg_broe,inplace=True)
```

```python
avg_stroke=df['stroke'].astype('float').mean(axis=0)
print('Average Broke',avg_stroke)
```

```python
df['stroke'].replace(np.nan,avg_stroke,inplace=True)
```

```python
avg_horsepower=df['horsepower'].astype(float).mean(axis=0)
print('Average horsepower',avg_horsepower)
```

```python
df['horsepower'].replace(np.nan,avg_horsepower,inplace=True)
```

```python
avg_rpm=df['peak-rpm'].astype('float').mean(axis=0)
print('Average peak_rpm',avg_rpm)
```

```python
df['peak-rpm'].replace(np.nan,avg_rpm,inplace=True)
```

```python
df['num-of-doors'].value_counts()
```

```python
df['num-of-doors'].replace(np.nan,"four",inplace=True)
```

```python
# simply drop whole row with NaN in "price" column
df.dropna(subset=["price"],axis=0,inplace=True)
```

```python
# reset index, because we droped two rows
df.reset_index(drop=True,inplace=True)
```
